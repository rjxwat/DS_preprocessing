```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#pd.pandas.set_option('display.max_columns',None)
```

```python
filepath="train.csv"
df=pd.read_csv(filepath)
df.shape
```

```python
df=df.iloc[:,1:27]
```

```python
df.info()
```

```python
df=df.dropna()
```

# Central tendancy

```python
df.mean()
```

```python
df.median()
```

```python
df.mode()
```

# Dispersion

```python
print('range\n',df.max()-df.min())
```

```python
IQR = df.quantile(0.75) - df.quantile(0.25)
print("Interquartile Range (IQR):\n", IQR)
```

```python
variance = df.var()

print("Variance of the dataset:\n", variance)
```

```python
std_dev = df.std()

print("Standard Deviation:\n", std_dev)
```

```python
mad =(df-df.mean()).abs()
print("Mean Absolute Deviation (MAD):\n", mad.mean())
```

# Shape analysis

```python
for column in df.columns:
    plt.figure()  # Create a new figure for each plot
    plt.hist(df[column])
    plt.title(column)  # Set title as the column name
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```

```python
part=df.iloc[:,:14]
```

```python
for column in part.columns:
    plt.figure()  # Create a new figure for each plot
    plt.boxplot(part[column])
    plt.title(column)  # Set title as the column name
    plt.ylabel("Value")
    plt.grid(True)
    plt.show()
```
